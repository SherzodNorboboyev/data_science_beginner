### exploratory data analysis

value_counts() = Frequency count

outliers = the value that  is considerably higher or lower from rest of the data

1. lets say value at 75% is Q3 and value 25% is Q1
2. outlier are smaller than Q1-1.5(Q3-Q1) and bigger than Q3+1.5(Q3-Q1). (Q3-Q1) = IQR<br>

We will use <b>describe()</b> method
<li>count = number of values</li>
<li>mean = avarage of values</li>
<li>std = standart devisasion</li>
<li>min = minimum of values</li>
<li>max = maximum of values</li>
<li>25% = first quantile</li>
<li>50% = median or second quantile</li>
<li>75% = third quantile</li>

what is quantile?

<li>1,4,5,6,8,9,11,12,13,14,15,16,17</li>
<li><b>median</b> is the number that is in the middle of the sequence. in this case it is 11</li>
<li><b>lower quartile</b> is the median in between the smallest number and the median i.e. in 1 and 11, which is 6</li>
<li><b>upper quartile</b> is the median in between the median and the largest number i.e. in 11 and 17, which is 14</li>


```python
# frequency
print(df['Type 1'].value_counts(dropna=False)) # nan also counted
```

### resampling pandas time series

* Resampling: statistical method over different time intervals
    * Needs string to specify frequency like 'M' = month or 'A' = year
* Downsampling: reduce date time rows to slower frequency like from daily to weekly
* Upsampling: Increase date  time rows to faster frequency like from daily to hourly
* Interpolate: interpolate values according to different methods like 'linear', 'time' or 'index'
    * https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.interpolate.html


```python
# resample with year
print(df1.resample('A').mean())
```

```python
# resample with month
print(df1.resample('M').mean())
# lot of NaN beacuse of there not so much months
```

```python
# in real life scenarios NaN would not be not so much, so we can fill them wih interpolate
# we can interpolate from first value
print(df1.resample('M').first().interpolate('linear'))

# or with mean()
print(df1.resample('M').mean().interpolate('linear'))
```

### pivoting
* pivoting - reshape tool

```python
dic = {'treatment':['A', 'A', 'B', 'B'], 'gender': ['F', 'M', 'F', 'M'], 'response': [10, 45, 5, 9], 'age': [15, 13, 72, 65]}
df = pd.DataFrame(dic)
print(df)
```

```python
# pivoting
print(df.pivot(index='treatment', columns='gender', values='response'))
```

### stacking and unstacking DF

* deal with multi label indexes
* level: position of unstacked index
* swaplevel: change inner and outer level index position

```python
# stack
df1 = df.set_index(['treatment','gender'])
print(df1)
```

```python
# unstack
# level determines indexes
print(df1.unstack(level=0))
print(df1.unstack(level=1))
```

```python
# change inner and outer level index position
df2 = df1.swaplevel(0,1)
print(df2)
```
